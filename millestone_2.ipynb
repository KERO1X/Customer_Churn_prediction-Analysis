```python
import numpy as np
import pandas as pd 
import seaborn as sns 
from matplotlib import pyplot as plt
df = pd.read_csv("/home/koks/Desktop/DEPI/final project/data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.describe()
df.head().T
```

```python
df.drop("customerID",axis=1,inplace=True)
df.head().T
```

```python
df.duplicated().sum()
```

```python
df.TotalCharges.describe()  #the program show object
df.TotalCharges.mode()[0]

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df[df['TotalCharges']==' '].T #the sol convert to numeric
```

```python

df.isnull().sum() #full missing values
```

```python
df['TotalCharges']=df['TotalCharges'].fillna(df['TotalCharges'].median())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pair Plot
sns.pairplot(df, hue='Churn', vars=['MonthlyCharges', 'tenure', 'TotalCharges'], height=3)
plt.suptitle("Pair Plot of Key Features", y=1.02)
plt.show()
```

```python
# Bar Chart for Categorical Features
categorical_features = ['gender', 'SeniorCitizen', 'Partner', 'Dependents']
for feature in categorical_features:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=feature, hue='Churn')
    plt.title(f'Churn Distribution by {feature.capitalize()}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
# Box Plot for Numerical Features
numerical_features = ['MonthlyCharges', 'tenure', 'TotalCharges']
for feature in numerical_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df, x='Churn', y=feature)
    plt.title(f'Distribution of {feature.capitalize()} by Churn')
    plt.show()
```

```python
# Pie Chart for Churn Rate
churn_counts = df['Churn'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(churn_counts, labels=['Non-Churn', 'Churn'], autopct='%1.1f%%', colors=['lightgreen', 'salmon'])
plt.title('Overall Churn Rate')
plt.show()
```

```python
# FacetGrid for Interaction Effects
g = sns.FacetGrid(df, col='Contract', row='PaymentMethod', hue='Churn')
g.map(sns.countplot, 'SeniorCitizen')
g.add_legend(title='Churn')
plt.suptitle('Interaction Between Contract, Payment Method, SeniorCitizen, and Churn')
plt.tight_layout()
plt.show()
```
